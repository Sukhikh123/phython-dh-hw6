# Первое задание

```python
def get_bigrams(string):
    bigrams = []
    
    for index in range(0, len(string) - 1):
        if string[index] != ' ' and string[index + 1] != ' ':
            bigrams.append(string[index] + string[index + 1])
    
    return bigrams


def get_bigram_counts(string):
    bigrams = get_bigrams(string)
    bigram_counts = {}
    
    for bigram in bigrams:
        if bigram not in bigram_counts:
            bigram_counts[bigram] = 0
        bigram_counts[bigram] += 1
        
    return bigram_counts


string = 'абракадабра фокус покус десятка за курс немедленно появись'

bigrams = get_bigrams(string)
for bigram in bigrams:
    print(bigram)
    
bigram_counts = get_bigram_counts(string)
for bigram, count in bigram_counts.items():
    if bigram[0] != bigram[1]:
        print(bigram, count)
```

# Второе задание

```python
def is_reversed(one, another):
    if len(one) != len(another):
        return False
    
    for index in range(len(one)):
        if one[index] != another[-index-1]:
            return False
    
    return True


def print_reversed(one, another):
    if is_reversed(one, another):
        print('YES')
    else:
        print('NO')


print_reversed('каша', 'ашак')
print_reversed('мыло', 'рама')
```

# Третье задание

```python
import random

SECRET_WORDS = [
    'стихотворение',
    'программа',
    'лотерея',
    'искусство',
    'компьютер'
]

ATTEMPTS = 15


def format_remaining_attempts(attempt):
    remainder_100 = attempt % 100
    remainder_10 = attempt % 10
    
    if (
        10 <= remainder_100 < 20 or
        5 <= remainder_10 <= 9 or
        remainder_10 == 0
    ):
        message = 'Осталось {} попыток'
    elif remainder_10 == 1:
        message = 'Осталась {} попытка'
    else:
        message = 'Осталось {} попытки'
    
    return message.format(attempt)

    
def is_letter(symbol):
    return len(symbol) == 1 and symbol.isalpha()

    
def response(symbol, correct_letters, previous_letters):
    if not is_letter(symbol):
        print('Нужно ввести одну букву')
        return
        
    if symbol in previous_letters:
        print('Вы уже называли эту букву')
        return
    
    if symbol in correct_letters:
        print('Вы угадали букву!')
    else:
        print('Вы не угадали букву...')
        

def update_candidate(candidate, word, symbol):
    for index in range(len(word)):
        if word[index] == symbol:
            candidate[index] = symbol


def play():
    secret_word = random.choice(SECRET_WORDS)
    
    correct_letters = set(secret_word)
    previous_letters = set()
    
    candidate_word = ['_'] * len(secret_word)
    
    print(
        'У вас есть {}, чтобы угадать слово длины {}'
        .format(
            format_remaining_attempts(ATTEMPTS)[len('Осталось '):],
            len(secret_word)
        )
    )

    print(' '.join(candidate_word))
    print()
    
    attempt = ATTEMPTS
    while attempt > 0:
        symbol = input()
        response(symbol, correct_letters, previous_letters)
        
        if is_letter(symbol):
            if symbol not in previous_letters and symbol in correct_letters:
                update_candidate(candidate_word, secret_word, symbol)
            
            if symbol not in previous_letters:
                previous_letters.add(symbol)
            
            print(' '.join(candidate_word))
        
        if candidate_word.count('_') == 0:
            print('Вы выиграли!')
            break
        
        attempt -= 1
        if attempt == 0:
            print('Вы проиграли...')
            break

        print(format_remaining_attempts(attempt))
        print()

play()
```
